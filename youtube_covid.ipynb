## Importando as bibliotecas

```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
```

## Abrindo a pasta dos arquivos baixados

```python
caminho_covid = 'C:/Users/joaovitor/Desktop/Dados'

lista = os.listdir(caminho_covid)
print(len(lista))
```

## Dando uma olhada nos arquivos presentes na pasta

```python
for i in lista:
    print(i) #queremos apenas os arquivos .shp , sem o .xml no final
```

## Pegando apenas os arquivos .shp da minha pasta

```python
lista_shp = []
for i in lista:
    if '.shp' in i:
        if '.xml' in i:
            pass
        else:
            lista_shp.append(i)
print(lista_shp)
print('==================')
print(len(lista_shp))
```

## Visualizando a tabela com os dados para apenas 1 dia, para exemplificar

```python
tabela_exemplo = gpd.read_file(os.path.join(caminho_covid, lista_shp[40]))
tabela_exemplo.head()
```

## Exibindo os dados geográficos

```python
tabela_exemplo.plot(figsize=(15,10), color='red', markersize=15)
```

## Importando o mapa do Brasil

```python
caminho_brasil = 'C:/Users/joaovitor/Desktop/bra_adm1/BRA_adm1.shp'
brasil = gpd.read_file(caminho_brasil)
brasil.head()
```

## Exibindo o mapa importado

```python
brasil.plot(figsize=(15,10), edgecolor='white', linewidth=0.5, color='#CCCCCC')
```

## Checando o crs de ambos os geodataframes

```python
print('brasil:', brasil.crs)
print('tabela_exemplo:', tabela_exemplo.crs)
```

## Igualando ambos os crs

```python
tabela_exemplo.crs = brasil.crs
```

## Ilustrando a sobreposição de mapas

```python
base = brasil.plot(figsize=(15,10), edgecolor='white', linewidth=0.5, color='#CCCCCC')

mapa_exemplo = tabela_exemplo.plot(ax=base, color='red')
mapa_exemplo.axis('off')
```

## Gerando vários mapas

```python
for key, values in enumerate(lista_shp):
    covid = gpd.read_file(os.path.join(caminho_covid, values))
    covid.crs = brasil.crs
    
    base = brasil.plot(color='#CCCCCC', edgecolor='white', linewidth=0.5, figsize=(15,10))
    ax = covid.plot(ax=base, color='red', markersize=13)
    ax.axis('off')
#===================================================================================
    ax.set_title(f'Disseminação COVID-19\n dia {values[9:-4]}-{values[6:8]}', \
                fontdict={'fontsize':'30', 'fontweight':'40'})
    ax.annotate('Para acessar os dados: www.ufrgs.br/sig/downloads/covid19-dados/\n Código disponível em github.com/joaotinti75',
               xy=(0.1, .08), xycoords='figure fraction', horizontalalignment='left', 
               verticalalignment='top', fontsize=20, color='#555555')
    base.set_axis_off()
    plt.savefig(f'C:/Users/joaovitor/Desktop/imagens/{key+1}.png')
    plt.close() #Para nao exibir as imagens no output, utilizo plt.close(). Consome menos memória
```

## Gerando o GIF

```python
import imageio

imagens = []
caminho_imagens = 'C:/Users/joaovitor/Desktop/imagens'
nome_arquivos = [f'{c}.png' for c in range(1, len(os.listdir(caminho_imagens))+1)]

for i in nome_arquivos:
    if i == '38.png': #caso a imagem 38 não tenha sido carregada com sucesso
        pass
    else:
        imagens.append(imageio.imread(os.path.join(caminho_imagens, i)))
    
imageio.mimsave('C:/Users/joaovitor/Desktop/mapa.gif', imagens, duration=0.1)
    
```
